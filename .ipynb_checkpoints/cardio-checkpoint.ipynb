```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cardio_preprocessed.csv',sep=',')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
## drop id colums
df.drop(columns=['id'],inplace=True)
df
```

```python
## converts age to years from days
df['age_years']=(df['age']/365).astype(int)
df.drop(columns=['age'],inplace=True)
df
```

```python
### calculate BMI
## BMI = weight / (height/100)^2
df['BMI']=df['weight'] / ((df['height']/100)**2)
df
```

```python
#give graphical represtation of null values
#EDA :- Exploratory Data Analysis
sns.heatmap(df.corr(), cmap='coolwarm',annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

<h1> Data Training </h1>

```python
!pip install scikit-learn
```

```python
from  sklearn.model_selection import train_test_split
## seperate input and output
## x = testing data
## y = target data

x = df.drop(columns=['cardio'],axis=1)  ## all data except cardio column is training data
y = df['cardio']  ## cardio is ourtarget data to check the correctness of model

x_train, x_test , y_train, y_test = train_test_split(
    x,y,test_size=0.2 , random_state=42,
)
```

```python
x_train
```

```python
df
```

```python
y = df['cardio']        # NOT scaled
y = y.astype(int)
```

```python
y = df['cardio'].astype(int)   # keep 0/1
X = df.drop('cardio', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

<h3>Logistic Regression</h3>

```python
### LogisticRegression

from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(
    max_iter=2000,
    C=5,                 # reduce regularization
    class_weight='balanced'
)
log_reg.fit(X_train, y_train)
```

```python
log_reg.score(x_test,y_test)*100
```

<h3> Decision Tree </h3>

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(
     max_depth=10,
    min_samples_split=5,
    random_state=42
)
dt.fit(x_train,y_train)
```

```python
dt.score(x_test,y_test)*100
```

<h3>Random forest</h3>

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(
     n_estimators=500,
    max_depth=12,
    class_weight='balanced',
    random_state=42
)
rf.fit(x_train,y_train)
```

```python
rf.score(x_test,y_test)*100
```

<h3>XGBOOST</h3>

```python
!pip install xgboost
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

xgb = XGBClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    use_label_encoder=False,
    eval_metric='logloss'
)

xgb.fit(x_train, y_train)
y_pred_xgb = xgb.predict(x_test)

accuracy_score(y_test, y_pred_xgb)*100
```

<h3>MODEL EVALUATION</h3>

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Predict on training and testing data
y_train_pred = xgb.predict(X_train)
y_test_pred = xgb.predict(X_test)

# Accuracy
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

print("Training Accuracy:", train_accuracy)
print("Testing Accuracy:", test_accuracy)

# Confusion Matrix
cm = confusion_matrix(y_test, y_test_pred)
print("\nConfusion Matrix:\n", cm)

# Classification Report
print("\nClassification Report:\n")
print(classification_report(y_test, y_test_pred))

# Overfitting / Underfitting check
if train_accuracy > test_accuracy:
    print("Model may be Overfitting")
elif train_accuracy < test_accuracy:
    print("Model may be Underfitting")
else:
    print("Model is well fitted")
```

<h3>CROSS VALIDATION</h3>

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(xgb, X_train, y_train, cv=5)

print("XGBoost CV Scores:", cv_scores)
print("Average CV Score:", cv_scores.mean())
```

<h3>Compare Multiple Models</h3>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "XGBoost": XGBClassifier(
        n_estimators=100,
        max_depth=5,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        use_label_encoder=False,
        eval_metric='logloss'
    )
}

for name, mdl in models.items():
    mdl.fit(X_train, y_train)
    score = mdl.score(X_test, y_test)
    print(f"{name} Accuracy: {score}")
```

<h3>Hyperparameter Tuning</h3>


```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [50, 100],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.05, 0.1]
}

XGBClassifier(
    random_state=42,
    eval_metric='logloss'
)
grid = GridSearchCV(
    xgb,
    param_grid,
    cv=5,
    scoring='accuracy'
)

grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best CV Score:", grid.best_score_)

```

<h3>Confusion Matrix Heatmap</h3>

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("XGBoost Confusion Matrix")
plt.show()
```

<h3>Accuracy Comparison Graph</h3>

```python
model_names = []
accuracies = []

for name, mdl in models.items():
    model_names.append(name)
    accuracies.append(mdl.score(X_test, y_test))

plt.figure(figsize=(8,5))
plt.bar(model_names, accuracies)
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.title("Model Accuracy Comparison (Including XGBoost)")
plt.show()
```
